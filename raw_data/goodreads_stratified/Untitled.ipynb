```python
srttrain = train[['user', 'movie']].sort_values(by=['user', 'movie'])
srttest = test[['user', 'movie']].sort_values(by=['user', 'movie'])

srtratingsout = ratingsout[['user', 'movie', 'rating']].sort_values(by=['user', 'movie'])
ratingstrain = []
i=0
for val in srtratingsout.iterrows():
#     print(val[1])
#     print(srt.iloc[i]['user']==val[1]['user'])
    if (srttrain.iloc[i]['user']==val[1]['user'] and  srttrain.iloc[i]['movie']==val[1]['movie']  ):
        ratingstrain.append(val[1]['rating'])
        i+=1

ratingstest = []
i=0
for val in srtratingsout.iterrows():
    if i < len(srttest) and (srttest.iloc[i]['user']==val[1]['user'] and  srttest.iloc[i]['movie']==val[1]['movie']  ):
        ratingstest.append(val[1]['rating'])
        i+=1
```

```python
srttrain['rating'] = ratingstrain
```

```python
srttest['rating'] = ratingstest
```
